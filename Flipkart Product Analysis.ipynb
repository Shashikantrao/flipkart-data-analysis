## Flipkart Product Analysis

### Import Libraries

```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use("default")
sns.set(style="whitegrid")
```

### Load Dataset

```python
df = pd.read_csv("flipkard.csv")
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df.describe(include="all").T
```

### Data Cleaning

```python
df.isnull().sum()
```

```python
# Fill numeric columns
num_cols = df.select_dtypes(include=np.number).columns
df[num_cols] = df[num_cols].fillna(df[num_cols].median())

# Fill categorical columns
cat_cols = df.select_dtypes(include="object").columns
df[cat_cols] = df[cat_cols].fillna("Unknown")
```

```python
df["listing_date"] = pd.to_datetime(df["listing_date"])
```

### Feature Engineering

```python
df["discount_amount"] = df["price"] - df["final_price"]

df["revenue"] = df["final_price"] * df["units_sold"]

df["stock_status"] = np.where(df["stock_available"] > 0, "In Stock", "Out of Stock")

df["listing_month"] = df["listing_date"].dt.month
df["listing_year"] = df["listing_date"].dt.year
```

### Overall Business Metrics (Numeric Insights)

```python
df[[
    "price","final_price","discount_percent","rating",
    "units_sold","revenue","review_count","seller_rating"
]].describe()
```

### Category-Level Analysis

```python
# Revenue by Category
category_revenue = df.groupby("category")["revenue"].sum().sort_values(ascending=False)
category_revenue.head(10)
```

```python
category_revenue.head(10).plot(kind="bar", figsize=(10,5), title="Top Categories by Revenue")
plt.ylabel("Revenue")
plt.show()
```

### Brand Performance Analysis

```python
top_brands = df.groupby("brand")["revenue"].sum().sort_values(ascending=False).head(10)
top_brands
```

```python
top_brands.plot(kind="barh", figsize=(10,5), title="Top Brands by Revenue")
plt.show()
```

### Price vs Rating Analysis

```python
sns.scatterplot(data=df, x="final_price", y="rating")
plt.title("Price vs Rating")
plt.show()
```

### Discount Impact on Sales

```python
sns.regplot(data=df, x="discount_percent", y="units_sold")
plt.title("Discount vs Units Sold")
plt.show()
```

### Seller Performance
